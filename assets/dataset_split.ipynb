```python
# imports 

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

TEST (HELDOUT)

```python
test = pd.read_csv('../data/heldout_HARDMath.csv') ## fix relative path

counts = test['question_type'].value_counts().sort_index()

# old_categories = ['nondimensionalization_numeric', 'nondimentionalization_numeric']
# new_category = 'nondimensionalization_numeric'

# if any(cat in counts.index for cat in old_categories):
#     counts[new_category] = counts[old_categories].sum()

# counts = counts.drop(['nondimentionalization_numeric'], errors='ignore')
```

```python
counts
```

```python
nondim_counts = counts[counts.index.str.contains('nondim', case=False)]
poly_counts = counts[counts.index.str.contains('polynomial', case=False)]
ode_counts = counts[counts.index.str.contains('ode', case=False)]
int_counts = counts[counts.index.str.contains('integral', case=False)]

labels = counts.index
sizes = counts.values
```

```python
fig, ax = plt.subplots()

cm = plt.get_cmap("tab20c")
type_colors = cm((np.arange(4)*4)+1)
subtype_colors = cm(np.array([2,6,10,11,14,15]))

ax.pie([ode_counts.sum(), int_counts.sum(), nondim_counts.sum(), poly_counts.sum()], 
       labels=['ODEs', 'Integrals', 'Nondim', 'Polynomials'], 
       colors=type_colors, autopct='%1.1f%%', startangle=90, counterclock=False, wedgeprops=dict(edgecolor='w'),
       labeldistance=0.3, pctdistance=0.52, rotatelabels=False)

# outer circle -> subtypes
ax.pie(sizes, 
       labels=['', '', 'Numeric\nNondim', 'Symbolic\nNondim', 'Root\nFinding', 'Root\nCorrection'], 
       radius=1, colors=subtype_colors, startangle=90, 
       counterclock=False, wedgeprops=dict(width=0.3, edgecolor='w'),
       labeldistance=0.8, rotatelabels=False)

plt.title('Problem Type Distribution in Heldout Set')
ax.axis('equal')
plt.show()
```
